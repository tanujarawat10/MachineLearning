```python
import numpy as np
import matplotlib.pyplot as plt
x=[1,5,8,10,14,18]
y=[1,1,10,20,4,75]
```

```python
#train algorithm
degree=2
poly_fit=np.poly1d(np.polyfit(x,y,degree))
```

```python
#plot data
xx=np.linspace(0,26,100)
plt.plot(xx,poly_fit(xx),c='r',linestyle='-')
plt.title('Polynimial')
plt.xlabel('X')
plt.ylabel('Y')
plt.axis([0,25,0,100])
plt.grid(True)
plt.scatter(x,y)
plt.show
```
